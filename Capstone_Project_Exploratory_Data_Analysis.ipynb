Write the summary here within 500-600 words.


The banking industry has seen a sharp increase in digital transactions over the past decade, bringing tremendous convenience to customers. However, this digital revolution has also opened doors for fraudsters who exploit vulnerabilities in online banking systems. This project focused on conducting an in-depth analysis of banking transaction data provided by Paisa Bazaar to detect fraudulent activities, understand key fraud patterns, and provide recommendations to strengthen fraud prevention mechanisms.

## Objective of the Project:

The primary objective was to analyze historical transaction data to identify suspicious and fraudulent patterns, understand the underlying factors that contribute to fraud, and develop predictive models that could detect fraud with high accuracy. The end goal was to help Paisa Bazaar implement automated solutions that minimize the risk of financial losses and enhance customer security.

 ## Data Collection and Preparation:

The dataset comprised several hundred thousand transaction records, containing fields such as customer ID, account details, transaction amount, transaction type, timestamp, device ID, and a label indicating whether the transaction was fraudulent. A key challenge was that fraudulent transactions made up less than 1% of the total dataset, making the data highly imbalanced.

Data preprocessing was an essential first step. We performed the following actions:

Handling Missing Values: Null and inconsistent entries were removed or imputed to avoid bias in the analysis.

Encoding Categorical Data: Fields such as transaction type and device ID were encoded using one-hot encoding to convert them into numerical form for model training.

Normalization: Continuous variables such as transaction amount were normalized to standardize the scale and avoid skewed results.

Outlier Treatment: Extreme values, especially in transaction amounts, were carefully handled to avoid distorting the model.

 ## Exploratory Data Analysis (EDA):

Through EDA, important insights were uncovered:

Fraudulent transactions tended to occur during non-standard banking hours, especially between 12 AM and 5 AM.

The average transaction amount in fraudulent cases was significantly higher than in legitimate ones, but there were many small fraud cases as well.

Certain transaction types, like large fund transfers or online bill payments, had a higher proportion of fraud compared to cash withdrawals.

Accounts with frequent small transactions over a short period indicated “smurfing,” a common tactic used to evade detection.

Visualizations like histograms and heatmaps helped highlight these patterns, and correlation analysis showed that features like transaction time and transaction type were strong indicators of fraud.

## Modeling Approach for Fraud Analysis

After completing the Exploratory Data Analysis (EDA), we focused on building machine learning models to detect fraudulent transactions automatically. The approach was kept simple and easy to understand.

 Step 1 – Data Preparation for Modeling

We used the cleaned dataset from EDA, where missing values were handled, categorical data was converted into numbers, and features were normalized.

Since fraud cases were very rare (less than 1% of all transactions), we used SMOTE (Synthetic Minority Over-sampling Technique) to balance the data by creating synthetic fraud examples. This helped the model learn patterns from both fraud and non-fraud transactions equally.

 Step 2 – Choosing Simple Models

We used three basic machine learning models:

Logistic Regression:

A simple model that predicts the probability of a transaction being fraud based on input features.

Random Forest Classifier:

A model that builds many decision trees and combines their results. It helps handle complex relationships in the data.

XGBoost Classifier:

An advanced model that focuses on improving predictions by correcting mistakes of previous models. It showed the best performance.

 Step 3 – Model Evaluation

To check how well the models were working, we used simple evaluation metrics:

Accuracy: How many transactions the model classified correctly.

Precision: Of the transactions flagged as fraud, how many were actually fraud.

Recall: Of all real fraud cases, how many were detected by the model (important because missing fraud cases is dangerous).

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Replace 'NM' in 'Payment_of_Min_Amount' with a more appropriate value (e.g., mode or a new category)
# Let's check the value counts first to decide
print(data['Payment_of_Min_Amount'].value_counts())

# Based on the counts, 'No' is the most frequent, so we can replace 'NM' with 'No' or treat it as a separate category.
# For now, let's replace 'NM' with 'No' assuming it means minimum payment was not made.
data['Payment_of_Min_Amount'] = data['Payment_of_Min_Amount'].replace('NM', 'No')

# Check the unique values again to confirm the change
print("\nUnique values in 'Payment_of_Min_Amount' after handling 'NM':")
print(data['Payment_of_Min_Amount'].unique())

# Explore the 'Type_of_Loan' column
print("\nUnique values in 'Type_of_Loan':")
print(data['Type_of_Loan'].unique())
print(f"Number of unique values in 'Type_of_Loan': {data['Type_of_Loan'].nunique()}")

# Explore the 'Credit_Score' column and map it to numerical values
print("\nUnique values in 'Credit_Score':")
print(data['Credit_Score'].unique())

# Map Credit_Score to numerical values (e.g., Good: 2, Standard: 1, Poor: 0)
credit_score_mapping = {'Good': 2, 'Standard': 1, 'Poor': 0}
data['Credit_Score_Encoded'] = data['Credit_Score'].map(credit_score_mapping)

# Display the first few rows with the new encoded column
display(data[['Credit_Score', 'Credit_Score_Encoded']].head())
````

Analyze the unique values in the 'Type_of_Loan' column and extract individual loan types to handle high cardinality.

````python
# Analyze unique values in 'Type_of_Loan'
print("\nValue counts for 'Type_of_Loan' (top 20):")
print(data['Type_of_Loan'].value_counts().head(20))

# Extract individual loan types
# First, replace the ' and ' separator with a comma for easier splitting
data['Type_of_Loan_cleaned'] = data['Type_of_Loan'].str.replace(' and ', ', ')

# Get all unique loan types by splitting the strings and flattening the list
all_loan_types = data['Type_of_Loan_cleaned'].str.split(', ').explode().unique()
print(f"\nAll individual loan types: {list(all_loan_types)}")

# Create binary columns for each individual loan type
for loan_type in all_loan_types:
    # Handle potential leading/trailing spaces from splitting
    loan_type_cleaned = loan_type.strip()
    if loan_type_cleaned and loan_type_cleaned != 'Not Specified': # Exclude empty strings and 'Not Specified'
        data[f'Has_{loan_type_cleaned.replace(" ", "_")}'] = data['Type_of_Loan_cleaned'].apply(lambda x: 1 if loan_type_cleaned in x else 0)

# Display the first few rows with the new binary columns
print("\nDataFrame with new binary loan type columns:")
display(data.head())

# Explore other categorical columns
print("\nUnique values and counts for other categorical columns:")
for col in ['Occupation', 'Credit_Mix', 'Payment_Behaviour']:
    print(f"\n'{col}':")
    print(data[col].value_counts())

# 'Credit_Mix' and 'Payment_Behaviour' seem suitable for one-hot encoding or label encoding depending on the model.
# 'Occupation' has 15 unique values, which is manageable for one-hot encoding if needed later.
````

Review the results of the previous step, specifically the extracted loan types, and refine the binary column creation to handle variations in loan type names and exclude irrelevant entries like 'No Data' and 'Not Specified'. Then, finalize the approach for encoding other categorical variables.

````python
# Review the identified individual loan types and clean them up
# It seems there are some loan types with trailing commas and 'Not Specified' entries.
# Let's create a cleaner list of relevant individual loan types.
relevant_loan_types = [
    'Auto Loan', 'Credit-Builder Loan', 'Personal Loan', 'Home Equity Loan',
    'Mortgage Loan', 'Student Loan', 'Payday Loan', 'Debt Consolidation Loan'
]

# Recreate binary columns based on the refined list, ensuring no duplicates or irrelevant columns
# Drop the previously created binary columns first to avoid duplicates
columns_to_drop = [col for col in data.columns if col.startswith('Has_')]
data.drop(columns=columns_to_drop, inplace=True)

# Recreate binary columns for each relevant individual loan type
for loan_type in relevant_loan_types:
    # Use the original 'Type_of_Loan' column for checking to avoid issues with the cleaned column
    data[f'Has_{loan_type.replace(" ", "_")}'] = data['Type_of_Loan'].apply(lambda x: 1 if loan_type in x else 0)

# Display the first few rows with the refined binary columns
print("\nDataFrame with refined binary loan type columns:")
display(data.head())

# Other categorical columns ('Occupation', 'Credit_Mix', 'Payment_Behaviour') were explored and seem manageable.
# 'Credit_Mix' and 'Credit_Score' have a natural order, so label encoding is suitable.
# 'Occupation' and 'Payment_Behaviour' do not have a natural order, so one-hot encoding would be appropriate if needed for modeling.
# For the purpose of EDA, we will keep them as they are for now, but note the potential encoding strategies for future steps.

# The 'Payment_of_Min_Amount' column was handled in the previous step by replacing 'NM' with 'No'.
# This column can be label encoded (Yes/No).

# The 'Name', 'ID', and 'Customer_ID' columns are identifiers and not useful for direct analysis of credit risk factors.
# 'SSN' is also an identifier and should not be used directly due to privacy concerns. These can be excluded from further analysis.

# The 'Type_of_Loan' and 'Type_of_Loan_cleaned' columns can be dropped after creating the binary columns.
data.drop(columns=['Type_of_Loan', 'Type_of_Loan_cleaned'], inplace=True)

print("\nDataFrame after dropping original loan type columns:")
display(data.head())
````
